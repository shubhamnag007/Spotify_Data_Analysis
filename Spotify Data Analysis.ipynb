```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading dataset
df_track = pd.read_csv('SpotifyFeatures.csv')
df_track.head()
```

```python
df_track.describe()
```

```python
#Checking datatypes
df_track.info()
```

```python
df_track.isnull().sum()
```

```python
#Checking top 10 most popular songs
df_most_popular_sorted = df_track.sort_values('popularity', ascending = False).head(10)
df_most_popular_sorted
```

```python
# Checking top 10 Least popular songs
df_least_popular_sorted = df_track.sort_values('popularity', ascending = True).head(10)
df_least_popular_sorted
```

```python
most_popular = df_track.query('popularity > 90', inplace= False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_track['duration'] = df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop('duration_ms', inplace= True, axis = 1)
```

```python
df_track.head()
```

```python
corr_df = df_track.drop(['key', 'mode'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = [14, 6])
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linecolor = 'Black')
heatmap.set_title('correlation heatmap between variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_df = df_track.sample(int(0.004 * len(df_track)))
```

```python
plt.figure(figsize = [10,6])
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c')
```

```python
plt.figure(figsize = [10,6])
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'g')
```
